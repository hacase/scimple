```python
import sys,time,random

string = """Show
Cinemas
In
My
Place
Listing
Everything"""

typing_speed = 100 #wpm

def slow_type(t):
    for l in t:
        sys.stdout.write(l)
        sys.stdout.flush()
        time.sleep(random.random()*10.0/typing_speed)
    print('')

slow_type(string)
```

```python
from urllib.request import urlopen
import re
from bs4 import BeautifulSoup

date = []
title = []
spec = []
time = []
location = []
ticket = []

url = "https://www.bonnerkinemathek.de/programm/"
page = urlopen(url)
html = page.read().decode("utf-8")

pattern = r'<h2>.*?</a></div>'
match_results = re.findall(pattern, html, re.DOTALL)

for i in match_results:    
    soup = BeautifulSoup(i, "html.parser")
    
    for i in range(100):
        try:
            date.append(re.sub("<.*?>", "", str(soup.find("h2"))))
            title.append(re.sub("<.*?>", "", soup.find_all("div", class_="title")[i].text))
            spec.append(re.sub("<.*?>", "", soup.find_all("span", class_="spec")[i].text))
            time.append(re.sub("<.*?>", "", soup.find_all("span", class_="date")[i].text))
            location.append(re.sub("<.*?>", "", soup.find_all("div", class_="location")[i].text))
            ticket.append(re.sub("<.*?>", "", soup.find_all("a", class_="movie block-1", href=True)[i].text))
        except:
            break
            
print(date[i])
for i in range(len(title)):
    if (date[i] != date[i+1]) or (i+1 == len(date)):
        print(date[i])

    print(title[i])
    print(ticket[i])
    print(spec[i], time[i])
    print(location[i], "\n")


print('done')
```

```python
from urllib.request import urlopen
import re
from bs4 import BeautifulSoup

date = []
title = []
spec = []
time = []
location = []
ticket = []

url = "https://www.bonnerkinemathek.de/programm/"
page = urlopen(url)
html = page.read().decode("utf-8")

pattern = r'<h2>.*?</a></div>'
match_results = re.findall(pattern, html, re.DOTALL)

for i in match_results:    
    soup = BeautifulSoup(i, "html.parser")
    
    for i in range(100):
        try:
            title.append(re.sub("<.*?>", "", soup.find_all("div", class_="title")[i].text))
            date.append(re.sub("<.*?>", "", str(soup.find("h2"))))
            spec.append(re.sub("<.*?>", "", soup.find_all("span", class_="spec")[i].text))
            time.append(re.sub("<.*?>", "", soup.find_all("span", class_="date")[i].text))
            location.append(re.sub("<.*?>", "", soup.find_all("div", class_="location")[i].text))
            ticket.append(re.sub("<.*?>", "", soup.find_all("a", class_="movie block-1", href=True)[i]['href']))
        except:
            break
            
for i in range(len(title)):
    if (date[i] != date[i-1]) or (i+1 == len(date)):
        print('-'*40)
        print(date[i])
        print('-'*40)

    print(title[i])
    print(ticket[i])
    print(spec[i], time[i])
    print(location[i], "\n")


print('done')
```
